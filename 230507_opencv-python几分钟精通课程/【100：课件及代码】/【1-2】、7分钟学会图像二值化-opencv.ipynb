# 四、二值化-基本效果

ret, dst = cv2.threshold(src, thresh, maxval, type)

- src： 输入图，只能输入单通道图像，通常来说为灰度图
- thresh： 阈值
- maxval： 当像素值超过了阈值（或者小于阈值，根据type来决定），所赋予的值
- type：二值化操作的类型，包含以下5种类型： 
- dst： 输出图


```python
- cv2.THRESH_BINARY        
大于阈值部分取maxval（最大值），小于阈值的取0
- cv2.THRESH_BINARY_INV    
THRESH_BINARY的反转
- cv2.THRESH_TRUNC         
大于阈值部分设为阈值，小于阈值的不变
- cv2.THRESH_TOZERO        
大于阈值的不变，小于阈值的全为0
- cv2.THRESH_TOZERO_INV    
THRESH_TOZERO的反转
```

````python
ret, img_binary= cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)
````

# 五、其它几种情况

```
- cv2.THRESH_BINARY        
大于阈值部分取maxval（最大值），小于阈值的取0
- cv2.THRESH_BINARY_INV    
THRESH_BINARY的反转
- cv2.THRESH_TRUNC         
大于阈值部分设为阈值，小于阈值的不变
- cv2.THRESH_TOZERO        
大于阈值的不变，小于阈值的全为0
- cv2.THRESH_TOZERO_INV    
THRESH_TOZERO的反转
```

````python
ret, img_binary= cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)
ret, img_binary_INV= cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY_INV)
ret, img_binary_TRUNC= cv2.threshold(img_gray,127,255,cv2.THRESH_TRUNC)
ret, img_binary_TOZERO= cv2.threshold(img_gray,127,255,cv2.THRESH_TOZERO)
ret, img_binary_TOZERO_INV= cv2.threshold(img_gray,127,255,cv2.THRESH_TOZERO_INV)


print("np.max(img_binary_TRUNC): {}".format(np.max(img_binary_TRUNC)))
````

````python
img_list=[
    img_gray,img_binary,
    img_binary_INV,img_binary_TRUNC,
    img_binary_TOZERO,img_binary_TOZERO_INV]
FryCvUtils.plt_show_img(img_list,3,[6,6])
````
