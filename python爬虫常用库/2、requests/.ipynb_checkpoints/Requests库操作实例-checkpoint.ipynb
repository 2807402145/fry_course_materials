## 普通get请求获取博客园博客数据

```python
import requests 
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36"
}
response=requests.get('https://www.cnblogs.com/Renyi-Fan/p/13257332.html',headers=headers)
print(response)
print(response.text)
```

## 利用cookie模拟登录，获取登录了才能看到的消息

```python
import requests 
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36"
}
response=requests.get('https://www.cnblogs.com/Renyi-Fan/p/13259490.html',headers=headers)
print(response)
print(response.text)
```

```python
# 获取cookie
response=requests.get("https://fanrenyi.com")
# print(response.cookies)
for key,value in response.cookies.items(): 
    print(key + '=' +value)
```

```python
# 获取cookie
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36"
}
response=requests.get("https://www.cnblogs.com/Renyi-Fan",headers=headers)
print(response.cookies)
for key,value in response.cookies.items(): 
    print(key + '=' +value)
```

```python
import requests 
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36"
}
cookies = {
    ".Cnblogs.AspNetCore.Cookies":"",
    "_gid":"GA1.2.1552723320.1594148642",
    "_ga":"GA1.2.1718512095.1585155221",}
response=requests.get('https://www.cnblogs.com/Renyi-Fan/p/13259490.html',headers=headers,cookies=cookies)
print(response)
print(response.text)
```
