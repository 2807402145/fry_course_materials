## 基础画图操作

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-1,1,50)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-1,1,50)
y = x**2
y2 = x
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

```python
import pandas as pd
import numpy as np
data = pd.DataFrame(np.random.randn(1000,4),index=np.arange(1000),columns=list("ABCD"))
data = data.cumsum()
print(data)
data.plot()
# plt.plot(data)
plt.show()
```
